```python
import sqlite3
import pandas as pd
import numpy as np
```

```python
# Create the connection
cnx = sqlite3.connect(r'bot_data_latest.db')

# create the dataframe from a query
df_videos = pd.read_sql_query("SELECT * FROM videos", cnx)
df_chnnels = pd.read_sql_query("SELECT * FROM channels", cnx)
```

```python
combined_df = df_videos.join(df_chnnels.set_index('channel_id')[['name', 'category']], on='channel_id', lsuffix="_video", rsuffix="_channel")
combined_df
```

### 1. For each channel, how many videos have they published per channel, when did they last publish and how many videos a month have they published on average

```python
# For each channel, how many videos have they published per channel

channel_video_df = combined_df[['video_id','name_channel']].copy()
channel_video_df.columns = channel_video_df.columns.str.replace('video_id', 'video_count')
channel_video_df.groupby(['name_channel']).count()
```

```python
# when did they last publish

channel_published_df = combined_df[['name_channel', 'published']].copy()
channel_published_df['published'] = channel_published_df['published'].max()
channel_published_df.groupby('name_channel').agg(lambda col: np.sort(col.unique()).tolist())

channel_published_df = channel_published_df.groupby('name_channel')['published'].apply(lambda x: list(np.unique(x)))
channel_published_df
```

### 2. For each category, which is the most popular day in a week to publish

```python
df2 = combined_df[['name_channel', 'published','category']].copy()
df2['published'] = pd.to_datetime(df2['published'])
df2['published'] = df2['published'].dt.day_name()
df2 = df2.groupby('category').agg(lambda col: np.sort(col.unique()).tolist())
df2['published']
```

### 3. For the top ten channels by number of videos this year, how many videos did they publish each month on average

```python
from datetime import *
dt = datetime.now()

df3 = combined_df[['name_channel', 'published','video_id', 'channel_id']].copy()
df3['published'] = pd.to_datetime(df3['published'])

df_current_year = df3[df3['published'].dt.year == dt.year]
df_current_year

channel_video_df = df_current_year[["video_id","name_channel", "published"]].copy()

channel_video_df['published'] = pd.to_datetime(channel_video_df['published'])

channels_df = channel_video_df.groupby(["name_channel", pd.Grouper(key='published', freq='M')]).count()
channels_df["video_id"].nlargest(n=10)
```
